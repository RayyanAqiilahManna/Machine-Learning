Prompt ChatGPT : beri kode untuk menampilkan boxplot dari fitur-fitur numerik

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Membaca file CSV dan membuat dataframe
df = pd.read_csv('/content/drive/MyDrive/Dataset/diamonds.csv')

# Menentukan ukuran font
plt.rc('font', size=11)

# Membuat boxplot untuk setiap fitur numerik
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include=['float64', 'int64']), color='lightyellow')
plt.title('Boxplot of Numeric Features', fontsize=14)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#SCATTER PLOT

#HEATMAP

Prompt ChatGPT : berikan kode untuk menampilkan heat map untuk semua fitur numerik

```python
# Membaca file CSV dan membuat dataframe
df = pd.read_csv('/content/drive/MyDrive/Dataset/diamonds.csv')

# Menghitung korelasi antara fitur numerik
correlation_matrix = df.select_dtypes(include=['float64', 'int64']).corr()

# Membuat heatmap dari korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numeric Features', fontsize=14)
plt.show()
```

#VIOLIN PLOT

Prompt ChatGPT : berikan kode untuk menampilkan violin plot untuk semua fitur numerik dengan tema warna coklat dengan font 11

```python
# Membaca file CSV dan membuat dataframe
df = pd.read_csv('/content/drive/MyDrive/Dataset/diamonds.csv')

# Menentukan ukuran font
plt.rc('font', size=11)

# Menentukan tema warna coklat
sns.set_palette("ch:s=.25,rot=-.25")

# Membuat violin plot untuk semua fitur numerik
plt.figure(figsize=(12, 8))
sns.violinplot(data=df.select_dtypes(include=['float64', 'int64']), inner='quartile')
plt.title('Violin Plot of Numeric Features', fontsize=14)
plt.xlabel('Features', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

#HISTOGRAM

Prompt ChatGPT : berikan kode untuk menampilkan histogram untuk semua fitur numerik dengan warna hijau

```python
# Membaca file CSV dan membuat dataframe
df = pd.read_csv('/content/drive/MyDrive/Dataset/diamonds.csv')

# Menentukan warna hijau
color = 'green'

# Menentukan ukuran font
plt.rc('font', size=11)

# Menampilkan histogram untuk semua fitur numerik
plt.figure(figsize=(12, 8))
df.select_dtypes(include=['float64', 'int64']).hist(color=color, alpha=0.7, bins=20, figsize=(10,6))
plt.suptitle('Histogram of Numeric Features', fontsize=14)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
