#FEATURE STATISTIC

Prompt ChatGPT : beri kode untuk membuat dataframe dari file csv yang terletak pada /content/drive/MyDrive/Dataset/laptop_price.csv

```python
import pandas as pd

# Membaca file CSV dan membuat dataframe dengan encoding 'latin1'
df = pd.read_csv('/content/drive/MyDrive/Dataset/laptop_price.csv', encoding='latin1')

# Menampilkan lima baris pertama dari dataframe
print(df.head())
```

```python
# Menampilkan statistik deskriptif dari semua fitur dalam dataframe
print(df.describe())
```

#BOX PLOT

```python
df.info()
```

Prompt ChatGPT : berikan kode untuk menampilkan boxplot dari semua fitur numerik dengan warna brown

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Memilih fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Membuat boxplot dari fitur numerik
plt.figure(figsize=(10, 6))
sns.boxplot(data=numerical_features, color='brown')
plt.title('Boxplot of Numerical Features', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#SCATTER PLOT

Prompt ChatGPT : berikan kode untuk menampilkan scatter plot untuk semua fitur numerik

```python
# Memilih fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Membuat scatter plot untuk semua fitur numerik
sns.pairplot(numerical_features)
plt.show()
```

#HEATMAP

Prompt ChatGPT : berikan kode untuk menampilkan heat map untuk semua fitur numerik

```python
# Memilih fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Menghitung korelasi antar fitur numerik
correlation_matrix = numerical_features.corr()

# Membuat heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```

#VIOLIN PLOT

Prompt ChatGPT : berikan kode untuk menampilkan violin plot untuk semua fitur numerik dengan tema warna biru dengan font 8

```python
# Mengatur tema seaborn
sns.set_theme(style="whitegrid", font_scale=0.8)

# Memilih fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Membuat violin plot untuk semua fitur numerik
plt.figure(figsize=(10, 6))
sns.violinplot(data=numerical_features, palette='Blues')
plt.title('Violin Plot of Numerical Features', fontsize=16)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

#HISTOGRAM

Prompt ChatGPT : berikan kode untuk menampilkan histogram untuk semua fitur numerik dengan warna pastel

```python
# Memilih fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Mengatur gaya seaborn menjadi pastel
sns.set_palette("pastel")

# Membuat histogram untuk semua fitur numerik
plt.figure(figsize=(10, 6))
numerical_features.hist(bins=20, edgecolor='black', linewidth=1.2)
plt.suptitle('Histogram of Numerical Features', fontsize=16)
plt.show()
```

#LINE PLOT

Prompt ChatGPT : berikan kode untuk menampilkan line plot untuk semua fitur numerik dengan font 12

```python
# Memilih fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64'])

# Mengatur ukuran font menjadi 12
plt.rcParams.update({'font.size': 12})

# Membuat line plot untuk semua fitur numerik
plt.figure(figsize=(10, 6))
for column in numerical_features.columns:
    plt.plot(df[column], label=column)

plt.title('Line Plot of Numerical Features', fontsize=16)
plt.xlabel('Index', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.legend(loc='upper right')
plt.grid(True)
plt.show()
```

#Scatter Matrix

Prompt ChatGPT : berikan kode untuk menampilkan contour plot untuk semua fitur numerik

```python
# Memilih dua fitur numerik untuk contour plot
numerical_features = df.select_dtypes(include=['float64', 'int64'])
feature1 = numerical_features.columns[0]
feature2 = numerical_features.columns[1]

# Membuat contour plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x=feature1, y=feature2, fill=True, cmap='viridis', levels=20)
plt.title('Contour Plot of {} vs {}'.format(feature1, feature2), fontsize=16)
plt.xlabel(feature1, fontsize=14)
plt.ylabel(feature2, fontsize=14)
plt.show()
```
